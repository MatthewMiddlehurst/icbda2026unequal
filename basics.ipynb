```python
from aeon.datasets import load_from_ts_file

train_X, train_y = load_from_ts_file("./data/ItalyPowerDemandTruncEnd_TRAIN.ts")
test_X, test_y = load_from_ts_file("./data/ItalyPowerDemandTruncEnd_TEST.ts")

print(type(train_X), type(train_X[0]))
print(type(train_y))

print(len(train_X), train_X[0].shape)
print(train_y.shape)
print(len(test_X), test_X[0].shape)
print(test_y.shape)
```

```python
from algorithm_code.hc2 import HIVECOTEV2

hc2 = HIVECOTEV2(random_state=0)
hc2.fit(train_X, train_y)
```

```python
preds = hc2.predict(test_X)
print(preds)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(test_y, preds)
print(f"Accuracy: {acc:.4f}")
```
